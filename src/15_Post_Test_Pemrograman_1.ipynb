## BAGIAN 5 — HIGHER ORDER FUNCTION
Modifikasi program pada bagian 4 untuk menambahkan HOF untuk:
* Menyaring judul yang masuk dalam kategori panjang
* Menampilkan judul dengan jumlah karakter terbesar (tanpa spasi)


```python
from functools import reduce

def clean_title(title):
    return title.strip().title()

def count_words(title):
    words = title.split()
    return len(words)

def count_characters_without_space(title):
    return len(title.replace(' ', ''))

category = lambda title: "Panjang" if count_words(title) >= 3 else "Pendek"

def long_title(title):
    return category(title) == "Panjang"

def longer_title(title1, title2):
    return title1 if count_characters_without_space(title1) > count_characters_without_space(title2) else title2

raw_inputs = input("Masukkan judul buku, pisahkan dengan | : ")
cleaned_titles = [clean_title(title) for title in raw_inputs.split('|')]
character_counts = [count_characters_without_space(title) for title in cleaned_titles]
title_word_counts_and_categories = {title: [count_words(title), category(title)] for title in cleaned_titles}

print(f'Panjang judul tanpa spasi: {character_counts}')
print(f'Jumlah kata dan kategori: {title_word_counts_and_categories}')

print(f'Judul dalam kategori panjang: {list(filter(long_title, cleaned_titles))}')
print(f'Judul dengan jumlah karakter terbesar (tanpa spasi): {reduce(longer_title, cleaned_titles)}')
```
